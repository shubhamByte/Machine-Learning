```python
import matplotlib.pyplot as plt
```

# Scatter

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x, y)
plt.show()
```

# Plot

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x2, y2, "g-.", linewidth = 5)

plt.plot(x, y)
plt.show()
```

```python
# plot function x^3
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3

# plt.plot(x, y, "b")   both ways
plt.plot(x, y, color = "black", marker = "+")
# marker is little different because it is over the line instead of removing it like "b+"

plt.show()
```

```python
# If only one list is present, it will be treated as y and natural numbers will be assumed as x
a = [3,4,5]
plt.plot(a)
plt.show()
```

# color, marker, linewidth, labels for axis, datasets and the title, axis, grid, text

```python
import numpy as np
# color, marker, linewidth, labels for axis, datasets and the title, axis, grid, text
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3
y2 = x**2
#plt.plot(x, y, color = "black", marker="p")
plt.plot(x, y, color = "black", linewidth = 5, label = "x^3")

plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

#plt.axis([0, 10, 0, 200])

plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2, 80, "Text", fontsize=12)

plt.legend()
plt.title("Matplotlib demo")
plt.show()
```

# Bubble Chart

```python
year = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
population = [100, 110, 120, 140, 150, 179, 187]
salary = [500, 700, 900, 1100, 1300, 1800, 2200]
```

```python
c = np.arange(len(year))

# s means size
plt.scatter(year, salary, s = salary, c = c, alpha = 0.9, edgecolor = 'b')
plt.scatter(year, salary, s = salary, c = c, alpha = 0.9, edgecolor = 'b', marker = '^')
plt.axis([2009,2017,0,2500])
plt.show()
```

# Pie Chart

```python
labels = ["A","B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["blue", "purple", "red", "pink"]
explode= [0.1, 0, 0, 0]
plt.pie(sizes, colors=colors, explode = explode, labels=labels, autopct="%.2f%%",  startangle=100)
plt.axis("equal")
plt.show()
```

Histogram is for frequency, bar graph is for x and y.

```python
a = [1,2,4,5,6,8,9,13,12,1,17,19,22,23,11,8]      # default 10 bins, width of a bin/column decided accordingly. may be in decimal.
plt.hist(a)
plt.show()
```

```python
plt.hist(a, edgecolor = 'black')
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
xt = np.arange(24)

plt.hist(a, bins = 22, edgecolor = 'black')
plt.xticks(xt)
plt.show()
```

```python
plt.hist(a, bins = [1,3,7,13,17,23], edgecolor = 'black')
plt.show()
```
