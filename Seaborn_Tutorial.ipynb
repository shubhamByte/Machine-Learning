```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
car_df = sns.load_dataset('car_crashes')
car_df.head()
```

# DISTRIBUTION PLOT

```python
# bins is like bucket.In how many buckets you want to divide your x-axis in?
# kde is kernel density estimation. It gives a smooth curve estimation.
# distplot is deprecated. New is histplot and displot.

sns.displot(car_df['not_distracted'],kde = False, bins = 25)
```

# JOINT PLOT

```python
sns.jointplot(x = 'speeding', y= 'alcohol', data = car_df, kind = 'hex')
```

# KDE

```python
# both the syntax is correct
# sns.kdeplot(x = 'alcohol', data = car_df)
sns.kdeplot(car_df['alcohol'])
```

#  Pair Plot

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(data = tips_df, hue = 'sex', palette = 'Blues')
```

# Style and Colors


```python
# size of the image
# facecolor -> outer color of the graph
plt.figure(figsize = (8,6), facecolor = 'w')

# style = one of {darkgrid, whitegrid, dark, white, ticks}
sns.set_style(style = 'ticks')

# The base context is "notebook", and the other contexts are "paper", "talk", and "poster
# font_scale
sns.set_context(context = 'paper',font_scale = 2)

sns.histplot(tips_df['tip'])

# for removing horizontal and vertical line
# first flot then only dispine work. moving this line above will not work.
sns.despine(left = True)
```

# Categorical Plots

### Bar Plots

```python
plt.figure(figsize=(8,6))   # this line should before plotting

# estimator value could be np.median, np.mean, np.var, np.cov
sns.barplot(x = 'sex', y = 'tip', data = tips_df, estimator= np.median)
```

# Count Plot

```python
# sns.countplot(x = tips_df['sex'])
# or
# similar to bar plot but here estimator is just counting the occurence

sns.countplot(x = 'sex', data = tips_df)
```

# Box Plot

```python
plt.figure(figsize=(8,6))
sns.boxplot(data = tips_df, x = 'day',y = 'total_bill', hue = 'sex')

plt.legend(loc = 'upper right')
```

# Violin Plot

```python
sns.violinplot(data = tips_df, x = 'day',y = 'total_bill', hue = 'sex')
```

```python
# split shows both hues together
sns.violinplot(data = tips_df, x = 'day',y = 'total_bill', hue = 'sex', split = True)
```

# Strip Plot

```python
plt.figure(figsize = (8,6))

# dodge makes different hue separate. jitter ensures different points are not overlayed (automatically set to true).

sns.stripplot(data = tips_df, x = 'day',y = 'total_bill', hue = 'sex',dodge = True, jitter = True)
```

# Palette

```python
sns.stripplot(data = tips_df, x = 'day',y = 'total_bill', hue = 'sex',dodge = True, jitter = True, palette = 'afmhot')
```
