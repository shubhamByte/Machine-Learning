```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Functional Plot

```python
x1 = np.linspace(0,10,20)
y1 = x1 ** 2

plt.plot(x1, y1)

plt.title('square function')
plt.xlabel('values of x')
plt.ylabel('x squared')

plt.show()
```

# Print Multiple Plots

### Subplot

```python

plt.subplot(2,2,1)
plt.plot(x1, x1**2, 'r')

plt.subplot(2, 2, 4)
plt.plot(x1, x1**0.5, 'b')

plt.show()
```

### Subplots

```python
# notice the difference. it is not subplot but subplots.

# You can use the subplots() function to create multiple plots in a single figure, which returns a figure object and an array of axes objects. 
# The number of rows and columns can be specified as arguments.

myfig, myaxes = plt.subplots(figsize=(8,6), nrows = 1, ncols = 3)

myaxes[2].set_title('1-x^2 function')
myaxes[2].plot(x1, 1- x1**2)

myaxes[0].set_title('blank graph')


plt.tight_layout()
plt.show()
```

# Using Figure Objects

```python
fig1 = plt.figure(figsize=(8,6))   # create a canvas/ blank space called figure

axes1 = fig1.add_axes([0.1,.1,0.9,0.9])   # create a graph(axes)

axes1.plot(x1, x1**2,'r', label = 'x-squared')  # create a function plot in a graph
axes1.plot(x1, x1**3, 'g', label = 'x-cubed')

axes1.set_xlabel('x-axis')
axes1.set_ylabel('y-axis')
axes1.set_title('functions plot')
axes1.legend()


axes2 = fig1.add_axes([0.5,.65,0.3,0.3])
axes2.plot(x1**2,x1,'b')
axes2.set_title('square root func..')
axes2.set_xlabel('x')
axes2.set_ylabel('y')

```

# Appearance Options

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])

# Default colors (b: blue, g: green, r: red, c: cyan, m: magenta,
# y: yellow, k: black, w: white)
# color="0.75" creates a 75% gray
# You can use hexcodes color="#eeefff"
# You can use color names found next like this color="burlywood"
# https://en.wikipedia.org/wiki/Web_colors
# alpha defines the percentage of opacity

# The default line width is 1, so to double it put in 2 and so forth

# There are many line styles 
# matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html
# You can also provide a sample like '-.'

# Markers can mark your provided points on the graph
# https://matplotlib.org/3.3.0/api/markers_api.html
# You can change the markersize as well

# markerfacecolor changes the marker fill color
# markeredgecolor changes the marker stroke color
# markeredgewidth changes the markers stroke size

axes_3.plot(x1,y1,color='navy', alpha=.75, lw=2, ls='-.',
           marker='o', markersize=7, markerfacecolor='y',
           markeredgecolor='y', markeredgewidth=4)

# Set the lower and upper bound of x & y axis
axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

# Add a grid, color, dashes(5pts 1 pt dashes separated by 2pt space)
axes_3.grid(True, color='0.6', dashes=(5, 2, 1, 2))
# Set grid background color
axes_3.set_facecolor('#FAEBD7')
```

# Save visualisation to a file

```python
# You can save your plots to numerous file types : png, pdf, ps, eps, svg, pgf, 
fig_3.savefig('1st_plot.png')
```
