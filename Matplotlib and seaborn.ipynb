```python
apart_df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv")
apart_df.head()
```

```python
# seeing the distribution of each of the column

# Plot 1
plt.figure(figsize = (14,6))
plt.subplot(1,2,1)
sns.histplot(data=apart_df, x='GrLivArea', alpha=0.9, bins=50)
plt.title("Ground living area distribution")
plt.xlabel('area')

# Plot 2
plt.subplot(1,2,2)

apart_df['SalePrice'] = apart_df['SalePrice']/1000

sns.histplot(data=apart_df, x='SalePrice', alpha=0.9, bins=50)
plt.title("Sale price distribution")
plt.xlabel('Sale price (k$)')

plt.show()
```

```python
# seeing the correlation between area and sale price using scattterplot

plt.figure(figsize=(8,6))

sns.scatterplot(data=apart_df, x='GrLivArea', y='SalePrice', alpha=0.8)

plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.title("Ground living area partially explains sale price of apartments")

plt.show()
```

```python
# Plot 1
plt.figure(figsize = (14,6))
plt.subplot(1,2,1)

sns.regplot(data=apart_df, x='GrLivArea', y='SalePrice', line_kws={"color":"k"})

plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.title("Linear Regression")


# Plot 2
plt.subplot(1,2,2)

sns.regplot(data=apart_df, x='GrLivArea', y='SalePrice', line_kws={"color":"k"}, lowess = True)

plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.title("Lowess")

plt.show()
```

```python
car_df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/6_SeveralNum.csv")
car_df
```

```python
data = car_df[['disp', 'drat', 'hp', 'mpg', 'qsec', 'wt']]
plt.figure(figsize = (18,10))

i = 1
for col in data.columns:
    plt.subplot(2,3,i)
    sns.histplot(data[col], kde=False, fill="#69b3a2", bins = 20)
    plt.title(col)
    i = i + 1
plt.show()
```

```python
sns.pairplot(data, vars=['mpg', 'disp', 'drat', 'hp', 'qsec', 'wt'], kind='reg', diag_kind='kde')
plt.tight_layout()
```

```python
plt.figure(figsize = (14,8))

from scipy.cluster.hierarchy import linkage, dendrogram, set_link_color_palette
from scipy.spatial.distance import pdist

# Compute pairwise distances
dist_matrix = pdist(data)

# Perform hierarchical clustering
linkage_matrix = linkage(dist_matrix, method='ward')

# Create dendrogram
dendrogram(linkage_matrix, color_threshold=350, orientation='left')

# Set colors of labels and branches based on clusters
set_link_color_palette(['#69b3a2', '#404080', 'orange'])

# Show the plot
plt.show()
```

```python
weapon_df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/7_OneCatOneNum.csv")
weapon_df.head()
```

```python
plt.figure(figsize = (12,10))

# filter out missing values
weapon_df = weapon_df[~weapon_df['Value'].isna()]

# sort by Value
weapon_df = weapon_df.sort_values(by='Value')

# make Country a categorical variable
weapon_df['Country'] = weapon_df['Country'].astype('category')

# The horizontal plot is made using the hline function
plt.hlines(y=weapon_df['Country'], xmin=0, xmax=weapon_df['Value'], color='skyblue')
plt.plot(weapon_df['Value'], weapon_df['Country'], "o")

sns.despine(left = True, bottom = True)
```

```python
plt.figure(figsize = (12,10))

plt.barh( weapon_df['Country'],weapon_df['Value'])

sns.despine(left = True, bottom = True)
```

```python
# set figure size
plt.figure(figsize=(20,10))


# plot polar axis
ax = plt.subplot(111, polar=True)

# remove grid
plt.axis('off')

# Set the coordinates limits
upperLimit = 20000
lowerLimit = 1000

# Compute max and min in the dataset
max = weapon_df['Value'].max()

slope = (max - lowerLimit) / max
heights = slope * weapon_df['Value'] + lowerLimit

# Compute the width of each bar. In total we have 2*Pi = 360
width = 2*np.pi / len(weapon_df['Country'])

# Compute the angle each bar is centered on:
indexes = list(range(1, len(weapon_df['Country'])+1))
angles = [element * width for element in indexes]
angles

# Draw bars
bars = ax.bar(
    x=angles, 
    height=heights, 
    width=width, 
    bottom=lowerLimit,
    linewidth=2, 
    edgecolor="white")

# little space between the bar and the label
labelPadding = 10

# Add labels
for bar, angle, height, label in zip(bars,angles, heights, weapon_df['Country']):

    # Labels are rotated. Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle)

    # Flip some labels upside down
    alignment = ""
    if angle >= np.pi/2 and angle < 3*np.pi/2:
        alignment = "right"
        rotation = rotation + 180
    else: 
        alignment = "left"

    # Finally add the labels
    ax.text(
        x=angle, 
        y=lowerLimit + bar.get_height() + labelPadding, 
        s=label, 
        ha=alignment, 
        va='center', 
        rotation=rotation, 
        rotation_mode="anchor") 
```

```python
rap_df = pd.read_csv('https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/14_SeveralIndepLists.csv',
                     error_bad_lines=False)

rap_df[['artist', 'song', 'word']] = rap_df['artist "song" "word"'].str.split(' ', expand=True)
rap_df.drop('artist "song" "word"', axis=1, inplace=True)

# Convert the object type column to string
rap_df['artist'] = rap_df['artist'].astype(str)
rap_df['word'] = rap_df['word'].astype(str)
rap_df['song'] = rap_df['song'].astype(str)

```

```python
wage_df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/9_OneNumSevCatSubgroupOneObs.csv")
wage_df.head()
```

```python
data = wage_df
plt.figure(figsize = (16,8))
# List of country with 4 values
with4 = data.groupby('Country').filter(lambda x: len(x) == 4)
plt.xticks(rotation=90)
# Grouped plot
sns.barplot(x='Country', y='Value', hue='TIME', data=with4, dodge=True)

plt.xlabel("Country")
plt.ylabel("gender wage gap (%)")
plt.legend(title="Time")
plt.title("")
plt.show()
```

```python
data = data[data['TIME'].isin([2000,2015])]
data = data.pivot(index='Country',columns='TIME',values='Value')
data = data[data[2000]!=-1]
data = data[data[2015]!=-1]
data["text"] = "Country: " + data.index + "<br>" + "Wage gap in 2000: " + data[2000].astype(str) + "%<br>Wage gap in 2015: " + data[2015].astype(str) + "%"
sns.scatterplot(x=2000, y=2015, data=data, s = 40, alpha = 0.5)
plt.xlabel("Gender wage gap in 2000 (%)")
plt.ylabel("Gender wage gap in 2015 (%)")
plt.title("")
plt.show()
```
